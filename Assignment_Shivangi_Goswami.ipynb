```python
from datetime import datetime

date_range = pd.date_range(start='1947-08-15', end=datetime.today())
df = pd.DataFrame({'Date': date_range})

df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

df = df[['Day', 'Month', 'Year']]

df
```

```python
def freq(item):
    return item[1]

def top_k(nums, k):
    fre_ = {}
    for num in nums:
        if num in fre_:
            fre_[num] += 1
        else:
            fre_[num] = 1
    sorted_freq = sorted(fre_.items(), key=freq, reverse=True)
    
    top_k_freq = [item[0] for item in sorted_freq[:k]]
    
    return top_k_freq

input_list = [1, 2, 3, 4, 5, 1, 2, 3, 1, 2, 1]


top_2_frequent = top_k(input_list, 2)
print("Top 2 most frequent items:", top_2_frequent)
```

```python
#5) Divide a student table having their marks into 4 equal marks groups (quartiles) and show their details.
```

```python
WITH Quartile_ AS (
    SELECT 
        student_name, 
        marks,
        (SELECT COUNT(*) FROM table) AS total_stuss,
        (SELECT PERCENTILE_CONT(0.25) WITHIN GROUP (ORDER BY marks) FROM table) AS quartile_1,
        (SELECT PERCENTILE_CONT(0.50) WITHIN GROUP (ORDER BY marks) FROM table) AS quartile_2,
        (SELECT PERCENTILE_CONT(0.75) WITHIN GROUP (ORDER BY marks) FROM table) AS quartile_3
    FROM table
)
SELECT 
    student_name, 
    marks,
    CASE 
        WHEN marks <= quartile_1 THEN 'Quartile 1'
        WHEN marks <= quartile_2 THEN 'Quartile 2'
        WHEN marks <= quartile_3 THEN 'Quartile 
        ELSE 'Quartile 4'
    END AS quartile
FROM Quartile_;
```
